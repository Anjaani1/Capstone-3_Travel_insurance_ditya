# Metric Evaluation
Dalam proyek ini, kita ingin membangun model yang dapat memprediksi apakah seorang pelanggan akan mengajukan klaim asuransi atau tidak. Untuk menilai seberapa baik model tersebut bekerja, kita perlu memahami beberapa istilah penting:
### Confusion Matrix

|                             | Prediksi: Klaim (1) | Prediksi: Tidak Klaim (0) |
|-----------------------------|---------------------|----------------------------|
| **Aktual: Klaim (1)**       | True Positive (TP)  | False Negative (FN)        |
| **Aktual: Tidak Klaim (0)** | False Positive (FP) | True Negative (TN)         |

- **True Positive (TP)**: Pelanggan mengajukan klaim dan model memprediksi klaim  Prediksi benar  
- **False Positive (FP)**: Pelanggan tidak klaim, tapi model memprediksi klaim  Dana disiapkan padahal tidak perlu  
- **False Negative (FN)**: Pelanggan mengajukan klaim, tapi model memprediksi tidak  Perusahaan tidak siap menanggung klaim  
- **True Negative (TN)**: Pelanggan tidak klaim dan model memprediksi tidak  Prediksi benar

### Target
1: Pemegang polis `mengajukan klaim` terhadap polis asuransi

0: Pemegang polis `tidak mengajukan klaim` terhadap polis asuransi

### Fokus Utama: Minimalkan False Negative

Karena kesalahan memprediksi pelanggan yang sebenarnya akan klaim (False Negative) bisa menyebabkan kerugian besar  seperti kekurangan dana dan reputasi buruk  maka metrik utama yang kita gunakan adalah:

- **Recall**: Seberapa banyak pelanggan yang benar-benar klaim berhasil dikenali model  
### Recall = $ \frac{TP}{TP + FN} $

Selain itu, kita juga mempertimbangkan:

- **ROC AUC (Receiver Operating Characteristic - Area Under Curve)**  
  Menilai keseimbangan model dalam membedakan antara klaim dan tidak klaim, cocok untuk data yang tidak seimbang.
  
- **Balanced Accuracy**  
  Digunakan sebagai pengganti accuracy biasa karena dataset memiliki jumlah pelanggan klaim yang jauh lebih sedikit dibanding yang tidak klaim.

---

### Model Selection

Beberapa model akan diuji menggunakan metrik di atas, dan model terbaik akan dipilih untuk dilakukan tuning lebih lanjut (hyperparameter tuning) demi mencapai performa optimal.

```python
cols = ['Duration', 'Net Sales', 'Commision (in value)', 'Age']

plt.figure(figsize=(14, 10))

for i, col in enumerate(cols, 1):
    plt.subplot(2, 2, i)
    sns.boxplot(data=df, x=col)
    
# Hitung outlier menggunakan metode IQR (hanya untuk teks, bukan filtering data)
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    outliers = df[(df[col] < lower_bound) | (df[col] > upper_bound)]
    
    plt.title(f"{col}\n{len(outliers)} outliers ({len(outliers)/len(df)*100:.2f}%)")

plt.tight_layout()
plt.show()
```

Analisis Outlier Melalui Boxplot

- **Duration:** Sebagian besar durasi berada pada angka yang rendah, tetapi ada beberapa nilai yang sangat tinggi. Hal ini menandakan adanya transaksi dengan durasi tidak biasa yang berpotensi sebagai outlier.

### - Modeling(Feature Selection & Feature Engineering)

```python
# Feature Selection
onehot_features = ["Agency","Agency Type","Distribution Channel","Product Name","Destination"]
numeric_features = ["Duration","Net Sales","Commision","Age"]

# Feature Engineering
selected_features = onehot_features+numeric_features
df[selected_features].head()
```

```python
# Pipeline untuk fitur kategori (one-hot encoding)
onehot_pipeline = Pipeline(
    [
        ("onehot", OneHotEncoder(handle_unknown="ignore", sparse_output=False)),
    ]
)


# Pipeline untuk fitur numerik
numeric_pipeline = Pipeline(
    [("scaler", RobustScaler())]
)

# Gabungkan semua pipeline ke dalam satu preprocessor berdasarkan tipe fitur
preprocessor = ColumnTransformer(
    [
        ("cat", onehot_pipeline, onehot_features),
        ("num", numeric_pipeline, numeric_features),
    ]
)

# Transformasi Fitur & Target
X = df_copy[onehot_features + numeric_features]
y =df_copy["Claim"]

# Transformasi fitur dengan preprocessor
X_preprocessed = preprocessor.fit_transform(X)
X_preprocessed=pd.DataFrame(X_preprocessed,columns=preprocessor.get_feature_names_out())
X_preprocessed
```

### - Modeling & Evaluation

```python
# Define model
logreg = LogisticRegression(max_iter=1000, random_state=42)
knn = KNeighborsClassifier()
tree = DecisionTreeClassifier(random_state=42)
rf = RandomForestClassifier(random_state=42)
ada = AdaBoostClassifier(random_state=42)
gbm = GradientBoostingClassifier(random_state=42)
lgbm = lgb.LGBMClassifier(random_state=42)
xgb = XGBClassifier(use_label_encoder=False, eval_metric='logloss', random_state=42)

# Define oversampling 
smote = SMOTE()
```

```python
# Split data terlebih dahulu (train = 80%, test = 20%)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, stratify=y, random_state=42)

# Skor khusus
precision_scorer = make_scorer(precision_score)
recall_scorer = make_scorer(recall_score)

# Strategi resampling
sampling_strategies = {
    "No Resampling": None,
    "SMOTE": SMOTE(random_state=42),
    "RandomOverSampler": RandomOverSampler(random_state=42),
    "RandomUnderSampler": RandomUnderSampler(random_state=42),
}

# Model
models = [logreg, knn, tree, rf, ada, xgb, lgbm, gbm]
model_names = [
    "Logistic Regression",
    "KNN",
    "Decision Tree",
    "Random Forest",
    "AdaBoost",
    "XGBoost",
    "LightGBM",
    "Gradient Boosting",
]

# Validasi silang di train set
skfold = StratifiedKFold(n_splits=5, shuffle=True, random_state=777)

# Semua hasil dikumpulkan di sini
all_results = []

for strategy_name, sampler in sampling_strategies.items():
    print(f"Evaluating with: {strategy_name}")

    mean_roc_auc, std_roc_auc = [], []
    mean_precision, std_precision = [], []
    mean_recall, std_recall = [], []

    for i in range(len(models)):
        steps = [("preprocessor", preprocessor)]
        if sampler is not None:
            steps.append(("sampler", sampler))
        steps.append(("classifier", models[i]))

        pipe = ImbPipeline(steps)

        scores = cross_validate(
            pipe,
            X_train,  # evaluasi hanya pada data latih
            y_train,
            cv=skfold,
            scoring={
                "roc_auc": "roc_auc",
                "precision": precision_scorer,
                "recall": recall_scorer,
            },
            return_train_score=False,
            error_score='raise'
        )

        mean_roc_auc.append(scores["test_roc_auc"].mean())
        std_roc_auc.append(scores["test_roc_auc"].std())
        mean_precision.append(scores["test_precision"].mean())
        std_precision.append(scores["test_precision"].std())
        mean_recall.append(scores["test_recall"].mean())
        std_recall.append(scores["test_recall"].std())

    # Simpan hasil dalam dataframe
    results = pd.DataFrame({
        "Sampling Strategy": strategy_name,
        "Model": model_names,
        "Mean ROC AUC": mean_roc_auc,
        "Std ROC AUC": std_roc_auc,
        "Mean Precision": mean_precision,
        "Std Precision": std_precision,
        "Mean Recall": mean_recall,
        "Std Recall": std_recall,
    })

    all_results.append(results)

# Gabungkan semua hasil jadi satu tabel besar
train_results = pd.concat(all_results, ignore_index=True)

# Urutkan (misalnya berdasarkan F2 Score)
train_results = train_results.sort_values(by=["Mean Recall"], ascending=[False])

# Tampilkan hasil
train_results
```

Tabel di atas adalah hasil evaluasi model klasifikasi dengan berbagai strategi sampling (penyeimbangan data) untuk data train, dilihat dari rata-rata performa (Mean ROC AUC, Precision, Recall) dan standar deviasi (Std) dari masing-masing metrik. Tujuannya ialah menentukan kombinasi model + sampling strategy terbaik berdasarkan performa stabil dan tinggi

Kesimpulannya:
- SMOTE + Logistic Regression memberi recall tertinggi (70.7%), cocok jika prioritas utama adalah mendeteksi semua kasus positif.
- SMOTE + Gradient Boosting memberikan performansi seimbang (ROC AUC tinggi dan recall baik).
- No Resampling menghasilkan performa terburuk: precision dan recall mendekati 0.

```python
# Split data terlebih dahulu (train = 80%, test = 20%)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, stratify=y, random_state=42)

# Menyimpan hasil evaluasi test set
test_results = []

# Loop setiap strategi sampling
for strategy_name, sampler in sampling_strategies.items():
    print(f"Evaluating TEST SET with: {strategy_name}")
    
    # Loop setiap model
    for model, model_name in zip(models, model_names):
        # Pipeline
        steps = [("preprocessor", preprocessor)]
        if sampler is not None:
            steps.append(("sampler", sampler))
        steps.append(("classifier", model))
        pipe = ImbPipeline(steps)

        # Fit pada TRAIN set
        pipe.fit(X_train, y_train)

        # Prediksi pada TEST set
        y_pred = pipe.predict(X_test)
        y_proba = pipe.predict_proba(X_test)[:, 1] if hasattr(pipe, "predict_proba") else None

        # Hitung metrik evaluasi
        roc_auc = roc_auc_score(y_test, y_proba) if y_proba is not None else None
        precision = precision_score(y_test, y_pred, zero_division=0)
        recall = recall_score(y_test, y_pred, zero_division=0)

        # Simpan hasilnya
        test_results.append({
            "Sampling Strategy": strategy_name,
            "Model": model_name,
            "ROC AUC": roc_auc,
            "Precision": precision,
            "Recall": recall
        })

# Gabungkan semua hasil ke dalam dataframe
test_results_df = pd.DataFrame(test_results)

# Urutkan berdasarkan Recall Score dalam setiap metode sampling
test_results_df = test_results_df.sort_values(by=["Recall"], ascending=[False])

# Tampilkan hasil
test_results_df
```

Tabel di atas menunjukkan perbandingan performa berbagai model machine learning dengan berbagai strategi penyeimbangan data (sampling) menggunakan data test berdasarkan tiga metrik utama:

- Precision: Ketepatan prediksi positif.
- Recall: Seberapa banyak kasus positif yang berhasil ditangkap.

2. Logistic Regression + RandomUnderSampler juga sangat baik:

3. Tanpa resampling (No Resampling) hasilnya sangat buruk:
   - Precision dan Recall = 0
   - Model gagal mendeteksi kasus positif karena data tidak seimbang.

Kesimpulan Singkat:

```python
# Pipeline
pipe = ImbPipeline([
    ("preprocessor", preprocessor),  # preprocessing kamu sendiri
    ("sampler", RandomUnderSampler(random_state=42)),
    ("classifier", KNeighborsClassifier())
])

# Parameter Grid
param_grid = {
    "sampler__sampling_strategy": [0.3, 0.5, 0.7, 1.0],  # banding kelas minoritas:mayoritas
    "classifier__n_neighbors": [3, 5, 7, 9],
    "classifier__weights": ['uniform', 'distance'],
    "classifier__p": [1, 2]  # 1 = Manhattan, 2 = Euclidean
}

# Stratified CV dan Scorer
cv = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)
scorer = make_scorer(recall_score)

# Grid Search
grid = GridSearchCV(
    pipe,
    param_grid,
    scoring=scorer,
    cv=cv,
    n_jobs=-1,
    verbose=1
)

# Fit ke training data
grid.fit(X_train, y_train)
```

```python
# Ambil model terbaik dari hasil grid search
best_model = grid.best_estimator_

# Prediksi untuk train dan test
y_pred_train = best_model.predict(X_train)
y_pred_test = best_model.predict(X_test)

# Hitung recall
train_recall = recall_score(y_train, y_pred_train, zero_division=0)
test_recall = recall_score(y_test, y_pred_test, zero_division=0)

# Tampilkan hasil recall
print(f"Recall (Train): {train_recall:.4f}")
print(f"Recall (Test) : {test_recall:.4f}")
```

```python
# 1. Pipeline KNN
pipe_knn = ImbPipeline([
    ("preprocessor", preprocessor),
    ("sampler", RandomUnderSampler(random_state=42)),
    ("classifier", KNeighborsClassifier())
])

# 2. Parameter Grid
param_grid_knn = {
    "sampler__sampling_strategy": [0.5, 0.7, 1.0],
    "classifier__n_neighbors": [3, 5, 7],
    "classifier__weights": ["uniform", "distance"],
    "classifier__p": [1, 2]  # 1 = Manhattan, 2 = Euclidean
}

# 3. Grid Search CV
grid_knn = GridSearchCV(
    pipe_knn,
    param_grid_knn,
    scoring=make_scorer(recall_score),
    cv=StratifiedKFold(n_splits=5, shuffle=True, random_state=42),
    n_jobs=-1,
    verbose=1
)

# 4. Fit ke data training
grid_knn.fit(X_train, y_train)

# 5. Ambil model terbaik
best_knn = grid_knn.best_estimator_

# 6. Evaluasi: Prediksi train dan test
y_pred_train = best_knn.predict(X_train)
y_pred_test = best_knn.predict(X_test)

# 7. Hitung recall
recall_train = recall_score(y_train, y_pred_train, zero_division=0)
recall_test = recall_score(y_test, y_pred_test, zero_division=0)

# 8. Tampilkan hasil
print(" Recall Hasil Tuning:")
print(f" Train Recall: {recall_train:.4f}")
print(f" Test  Recall: {recall_test:.4f}")
```
